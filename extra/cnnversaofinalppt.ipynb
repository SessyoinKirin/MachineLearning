```python
def VGG16():
   model = models.Sequential([

        layers.Rescaling(1./255, input_shape=(224,224,3)),
        # Arquitetura enxuta
        layers.Conv2D(32, (3,3), padding='same', activation='relu'),
        layers.Conv2D(32, (3,3), padding='same', activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.1),

        layers.Conv2D(64, (3,3), padding='same', activation='relu'),
        layers.Conv2D(64, (3,3), padding='same', activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.1),

        layers.GlobalAveragePooling2D(),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(1, activation='sigmoid')
   ])
   model.compile(optimizer=Adam(learning_rate=1e-3), loss='binary_crossentropy', metrics=['accuracy'])
#    Loss = -[y * log(p) + (1 - y) * log(1 - p)]
   return model
```

```python
def matplot_acc_loss(train_process):
    # 'train_process' é o objeto 'history' retornado pelo model.fit
    # Os dados estão no dicionário 'train_process.history'
    history_dict = train_process.history

    # Crie o eixo X (número de épocas)
    epochs_range = range(len(history_dict['loss']))

    plt.figure(figsize=(12,4))

    # --- Gráfico de Loss ---
    plt.subplot(1,2,1)
    # Acesse os dados do dicionário
    plt.plot(epochs_range, history_dict['loss'], 'ro-', label='Train Loss', linewidth=2)
    plt.plot(epochs_range, history_dict['val_loss'], 'bs-', label='Val Loss', linewidth=2)
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Loss', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)

    # --- Gráfico de Accuracy ---
    plt.subplot(1,2,2)
    # Use 'accuracy' e 'val_accuracy'
    plt.plot(epochs_range, history_dict['accuracy'], 'ro-', label='Train Acc', linewidth=2)
    plt.plot(epochs_range, history_dict['val_accuracy'], 'bs-', label='Val Acc', linewidth=2)
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Accuracy', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.show()
```

```python
# Função auxiliar para inspecionar os lotes de dados
def debug_data(dataset, message):
    print(f"\n--- DEBUG: {message} ---")
    # Pega um lote (batch) do dataset
    for images, labels in dataset.take(1):
        print(f"Formato (Shape) do lote de imagens: {images.shape}")
        print(f"Tipo de dados (dtype) das imagens: {images.dtype}")
        print(f"Valor MÍNIMO dos pixels: {tf.reduce_min(images).numpy()}")
        print(f"Valor MÁXIMO dos pixels: {tf.reduce_max(images).numpy()}")
        print(f"-----------------------------------\n")
```

```python
def create_all_datasets(datadir='/teamspace/studios/this_studio/dataset_rbmk', img_size=(224,224), batch_size=32, seed=1):

    # Carrega dados de 1 diretório e divide CORRETAMENTE em 3 conjuntos:
    # Treino (70%), Validação (15%) e Teste (15%).

    # Otimiza o pipeline. A NORMALIZAÇÃO FOI REMOVIDA
    # pois ela já está DENTRO do modelo (Rescaling Layer).

    print("--- Carregando dataset completo para divisão manual (70/15/15) ---")

    # 1. Carrega o dataset INTEIRO de uma vez
    full_ds = tf.keras.preprocessing.image_dataset_from_directory(
        datadir,
        seed=seed,
        shuffle=True,
        image_size=img_size,
        batch_size=None # Carrega imagens individuais para podermos dividir
    )

    class_names = full_ds.class_names
    print(f"Classes encontradas: {class_names}")


# --- 2. Define os tamanhos da divisão ---
    data_size = tf.data.experimental.cardinality(full_ds)

    # --- INÍCIO DA CORREÇÃO ---
    # Precisamos converter data_size para float32 ANTES de multiplicar
    data_size_float = tf.cast(data_size, tf.float32)

    # Agora, fazemos a multiplicação com floats (o resultado é um float32 Tensor)
    train_size_float = data_size_float * 0.7
    val_size_float = data_size_float * 0.15

    # Convertemos o resultado de volta para int64 (o tf.cast trunca a vírgula)
    train_size = tf.cast(train_size_float, tf.int64)
    val_size = tf.cast(val_size_float, tf.int64)
    # --- FIM DA CORREÇÃO ---

    # Esta linha agora funciona, pois é (int64_tensor - int64_tensor - int64_tensor)
    test_size = data_size - train_size - val_size

    print(f"Total de Imagens: {data_size}")
    print(f" -> Tamanho do Treino (70%): {train_size}")
    print(f" -> Tamanho da Validação (15%): {val_size}")
    print(f" -> Tamanho do Teste (15%): {test_size}")




    # --- 3. Divide o dataset usando .take() e .skip() ---
    train_ds = full_ds.take(train_size)
    val_ds = full_ds.skip(train_size).take(val_size)
    test_ds = full_ds.skip(train_size + val_size).take(test_size)

    data_augmentation = tf.keras.Sequential([
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.05, fill_mode='nearest'),
        layers.RandomZoom(0.05, fill_mode='nearest')
    ], name="data_augmentation")

    # --- 4. Define Otimização (SEM NORMALIZAÇÃO) ---
    def optimize_pipeline(ds, is_training=False):
        ds = ds.cache()

        # --- ADICIONE ESTA LINHA ---
        # Se for o dataset de treino, re-embaralhe a cada época
        if is_training:
            # buffer_size grande garante um bom embaralhamento.
            # Um valor igual ao train_size é perfeito, mas 1000 já é ótimo.
            ds = ds.shuffle(buffer_size=1000, seed=seed, reshuffle_each_iteration=True)
            ds = ds.map(lambda x, y: (data_augmentation(x, training=True), y),
                        num_parallel_calls=tf.data.AUTOTUNE)
        # --- FIM DA ADIÇÃO ---

        # (Data augmentation pode ser adicionada aqui)
        ds = ds.batch(batch_size)
        ds = ds.prefetch(tf.data.AUTOTUNE)
        return ds

    # Aplica a otimização em cada conjunto
    train_ds = optimize_pipeline(train_ds, is_training=True)
    val_ds = optimize_pipeline(val_ds, is_training=False)
    test_ds = optimize_pipeline(test_ds, is_training=False)

    print("\n--- Pipelines de Treino, Validação e Teste criados com sucesso! ---")

    # --- 5. ADICIONE O BLOCO DE DEBUG AQUI ---
    print("\n--- DEBUG: Checando o pipeline de TREINO (Valores devem ser 0-255) ---")
    try:
        # Pega o primeiro lote (batch) do dataset de treino
        for images_batch, labels_batch in train_ds.take(1):
            print(f"Shape do Lote (Batch): {images_batch.shape}")

            # Converte para NumPy para pegar min/max facilmente
            images_np = images_batch.numpy()
            print(f"Valores Min/Max do Lote: {np.min(images_np)} / {np.max(images_np)}")

            # Verifica o tipo de dados
            print(f"Tipo de dados (DType): {images_np.dtype}")

            # Imprime um pequeno canto da primeira imagem
            print("Canto da primeira imagem (raw):")
            print(images_np[0, 0:5, 0:5, 0]) # Pixels [0-4] x [0-4] do canal Vermelho

    except Exception as e:
        print(f"Erro durante o debug do dataset: {e}")
    print("--------------------------------------------------\n")
    # --- FIM DO BLOCO DE DEBUG ---

    return train_ds, val_ds, test_ds, class_names
```

```python
def train_model_process(datadir='/teamspace/studios/this_studio/dataset_rbmk', img_size=(224,224), batch_size=32, epochs=40, model=None, save_path='melhor.h5'):

    # 25/10
    # como clear e cloudy é desbalanceado em quantidade,
    # aproximadamente 75% - 25%, o programa começou a chutar,
    # resultando 75% de acurácia
    # resolvi adicionar os pesos:
    # total = 37781
    # count_class_0 = 37781 * 0,753 = 28849
    # count_class_1 = 37781 * 0,247 = 9332
    # weight_for_0 = (1 / count_class_0) * (total / 2) = 0,66
    # weight_for_1 = (1 / count_class_1) * (total / 2) = 2,02
    # fazendo isso, um erro na classe 1 (cloudy) terá um impacto
    # mais ou menos 3 vezes maior no loss do que um erro na classe 0 (clear)
    # isso força o modelo a parar de chutar e realmente aprender a identificar
    # as caracteristicas.
    # 26/10 : não funciona, continuou chutando
    # atualização, funciona, o que não funcionou foi usar imagem que não era RGB, saiu tudo (0, 0, 0)
    class_weights = {0: 0.66, 1: 2.02}

    print("--- INICIANDO train_model_process ---")
    print("Chamando 'create_all_datasets'...")

    # --- ALTERAÇÃO 1: Chame a nova função ---
    # Ela agora retorna 4 itens, incluindo o test_ds
    train_ds, val_ds, test_ds, class_names = create_all_datasets(
        datadir=datadir,
        img_size=img_size,
        batch_size=batch_size,
        seed=1
    )

    print(f"...Datasets recebidos. Nomes das classes: {class_names}")

    # cria o modelo uma vez (se não foi passado)
    if model is None:
        model = VGG16()
        model.summary()

    # callbacks Keras
    callbacks = [
        ModelCheckpoint(save_path, monitor='val_accuracy', save_best_only=True, mode='max'),
        ReduceLROnPlateau(monitor='val_loss', factor=0.25, patience=4, min_lr=1e-6),
        EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)
    ]




    # Treina com os datasets tf.data (não use ImageDataGenerator com tf.data)
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=epochs,
        callbacks=callbacks,
        #class_weight = class_weights,
        verbose=1
    )


    matplot_acc_loss(history)

    # --- ALTERAÇÃO 2: Coleta de rótulos do TEST_DS ---
    print("Coletando rótulos verdadeiros (y_true) do TEST_DS...")
    try:
        # Use o test_ds, não o val_ds!
        y_true = np.concatenate([y for x, y in test_ds], axis=0)
        print(f"Rótulos de teste coletados. Total: {len(y_true)}")
    except Exception as e:
        print(f"Erro ao extrair rótulos y_true: {e}")
        y_true = None

    # --- ALTERAÇÃO 3: Avalie no TEST_DS ---
    print('Passando por situação final (avaliação real) usando melhor.h5 no TEST_DS')

    y_pred_probs = None
    best = None

    try:
        best = tf.keras.models.load_model(save_path)

        # Use o test_ds, não o val_ds!
        loss, acc = best.evaluate(test_ds, verbose=1)
        print(f'Best model loaded -> test_loss: {loss}, test_acc: {acc}')

        print("Gerando predições (para relatório completo) com o 'best' model no TEST_DS...")
        # Use o test_ds, não o val_ds!
        y_pred_probs = best.predict(test_ds, verbose=1)

    except Exception as e:
        print(f'Could not load saved model ({e}); evaluating current model object.')
        # Se 'model' foi treinado, ele já tem os melhores pesos por causa do restore_best_weights=True
        best = model
        loss, acc = best.evaluate(test_ds, verbose=1)
        print(f'val_loss: {loss}, val_acc: {acc}')

        print("Gerando predições (para relatório completo) com o 'model' (objeto em memória)...")
        y_pred_probs = best.predict(test_ds, verbose=1)

    # Assumindo que o 'best' model foi carregado
    if best is not None and class_names is not None:
        print("\n" + "="*50)
        print("DEBUG DETALHADO: 5 Primeiras Imagens (BinaryCrossentropy)")
        print("="*50)
        try:
            # 1. Pega o primeiro lote de validação
            for images_batch, labels_batch in test_ds.take(1):
                n_images = 5
                images_sample = images_batch[:n_images]
                labels_sample = labels_batch[:n_images]

                # 2. Como o algoritmo "identifica" (Predição)
                print("\n--- 1. Identificação (Predição) ---")
                # A saída do modelo agora é um único número por imagem
                pred_probs_sample = best.predict(images_sample) # Saída ex: [0.95], [0.02], ...

                # Instancia a loss function.
                # from_logits=False pois sua camada final tem 'sigmoid'
                loss_fn = tf.keras.losses.BinaryCrossentropy(from_logits=False)

                for i in range(n_images):
                    true_label_int = labels_sample.numpy()[i]   # ex: 1
                    true_label_float = float(true_label_int)    # ex: 1.0
                    true_name = class_names[true_label_int]

                    # A saída do modelo é um array de 1 elemento, ex: [0.95]
                    # Pegamos o valor escalar, ex: 0.95
                    pred_prob_float = pred_probs_sample[i][0]

                    # Identificação:
                    pred_class_int = 1 if pred_prob_float > 0.5 else 0
                    pred_name = class_names[pred_class_int]

                    # 3. Como o "loss" é calculado
                    print("\n--- 2. Cálculo do Loss (Individual) ---")

                    # Calcula o loss para esta única imagem
                    individual_loss = loss_fn(labels_sample[i:i+1], pred_probs_sample[i:i+1])

                    # Explicação do cálculo do loss (Binary Cross-Entropy):
                    # Loss = -[y * log(p) + (1 - y) * log(1 - p)]
                    # (Adicionamos +1e-7 para evitar log(0))
                    p = pred_prob_float + 1e-7
                    y = true_label_float
                    calculo_manual = -(y * np.log(p) + (1 - y) * np.log(1 - p))

                    print(f"\nIMAGEM {i+1}:")
                    print(f"  > Rótulo Verdadeiro: {true_label_int} ({true_name})")
                    print(f"  > Predição:          {pred_class_int} ({pred_name})")
                    print(f"  > Saída Bruta (Sigmoid): {pred_prob_float:.4f} (Prob. de ser classe 1)")
                    print(f"  > Loss (via Keras): {individual_loss.numpy():.4f}")
                    print(f"  > Verificação Manual (Loss = -[{y}*log({p}) + (1-{y})*log(1-{p})]): {calculo_manual:.4f}")
                    print(f"  > Fim de Verificação Manual...")

        except Exception as e_debug:
            print(f"Erro durante o debug detalhado: {e_debug}")
        print("="*50 + "\n")
    # --- FIM DO NOVO BLOCO DE DEBUG ---

    # --- Relatórios de Classificação e Matriz de Confusão (como antes) ---
    if y_pred_probs is not None and y_true is not None:
        y_pred_classes = (y_pred_probs > 0.5).astype(int)

        print("\n" + "="*50)
        print("RELATÓRIO DE CLASSIFICAÇÃO (TEST SET)")
        print("="*50)
        print(classification_report(y_true, y_pred_classes, target_names=class_names))

        print("\n" + "="*50)
        print("MATRIZ DE CONFUSÃO (TEST SET)")
        print("="*50)
        print(confusion_matrix(y_true, y_pred_classes))

    return model, history
```
